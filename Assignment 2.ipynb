```python
columns = ['text']
L = []
```

```python
files = ['female1.txt','female2.txt','female3.txt','female4.txt','female5.txt','female6.txt','female7.txt','female8.txt','female9.txt', 'female10.txt', 'female11.txt','female12.txt']
```

```python
 for i in files:
    filepath = i
    with open(filepath) as fp:
        for cnt, line in enumerate(fp):
            L.append(line)
```

```python
femaleresults = pd.DataFrame(L, columns=columns)
femaleresults['goodness']=0.000
len(femaleresults)
```

```python
dropcount = 0
for ind, i in enumerate(femaleresults['text']):
    strt = i.find('She')
    if strt == -1:
        if i.find('\t') != -1:
            tab = i.find('\t')
            i = i[(tab+1):]
            femaleresults['text'][ind] = "She's " + i
        else:
            femaleresults = femaleresults.drop(femaleresults.index[ind])
            dropcount += 1
    elif strt != 0:
        new = i[strt:]
        femaleresults['text'][ind] = new
print(dropcount)
```

```python
for ind, i in enumerate(femaleresults['text']):
    end = i.find(".")
    if i.find(".") != -1:
        new = i[:end]
        femaleresults['text'][ind] = new
```

```python
for ind in femaleresults.index:
    sent = TextBlob(femaleresults['text'][ind])
    femaleresults['goodness'][ind]=sent.sentiment[0]
```

```python
from textblob.np_extractors import ConllExtractor
extractor = ConllExtractor()

import nltk
nltk.download('punkt')
nltk.download('conll2000')
```

```python
#Most common descriptors
from nltk.probability import FreqDist

descriptors = []
for i in maleresults['text']:
    phrase = TextBlob(i, np_extractor=extractor)
    phlist = phrase.noun_phrases
    for i in phlist:
        descriptors.append(i)
for i in femaleresults['text']:
    phrase = TextBlob(i, np_extractor=extractor)
    phlist = phrase.noun_phrases
    for i in phlist:
        descriptors.append(i)
        
freqm = FreqDist(descriptors)
freqm.most_common(10)
```
